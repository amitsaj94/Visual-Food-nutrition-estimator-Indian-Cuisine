```python
import pandas as pd

# Load the file 
df = pd.read_excel("Anuvaad_INDB_2024.11.xlsx") 
```

```python
df
```

```python
df.columns
```

```python
# Standardize column names
df.columns = [col.lower().strip().replace(" ", "_") for col in df.columns]
```

```python
df.columns
```

```python
# Step 1: Calculate serving size in grams (important for scaling)
df['serving_size_g'] = (df['unit_serving_energy_kj'] / df['energy_kj']) * 100
```

```python
# Step 2: Select important columns for the model
selected_columns = [
    'food_name',
    'unit_serving_energy_kcal',
    'unit_serving_carb_g',
    'unit_serving_protein_g',
    'unit_serving_fat_g',
    'serving_size_g',                  
    'unit_serving_freesugar_g',        
    'unit_serving_fibre_g',            
    'unit_serving_sodium_mg'           
]
```

```python
# Step 3: Filter DataFrame
formatted_df = df[selected_columns].copy()

# Rename columns for model clarity
formatted_df = formatted_df.rename(columns={
    'food_name': 'food',
    'unit_serving_energy_kcal': 'calories_kcal',
    'unit_serving_carb_g': 'carbs_g',
    'unit_serving_protein_g': 'protein_g',
    'unit_serving_fat_g': 'fat_g',
    'unit_serving_freesugar_g': 'sugar_g',
    'unit_serving_fibre_g': 'fibre_g',
    'unit_serving_sodium_mg': 'sodium_mg'
})
```

```python
formatted_df.head()
```

```python
formatted_df
```

```python

essential_columns = [
    'calories_kcal', 'carbs_g', 'protein_g', 'fat_g',
    'serving_size_g', 'sugar_g', 'fibre_g', 'sodium_mg'
]

# Drop rows with missing values in any essential column
cleaned_df = formatted_df.dropna(subset=essential_columns).copy()

# Reset index after dropping rows
cleaned_df.reset_index(drop=True, inplace=True)

# Show how many rows earlier and now remained after cleaning
print(f"Original dataset size: {len(formatted_df)} rows")

print(f"Cleaned dataset size: {len(cleaned_df)} rows")

```
